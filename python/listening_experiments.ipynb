```python
def sin_freq_ms(freq, time_in_ms, sample_rate_hz=44100):
  return np.sin(
    np.arange(time_in_ms*sample_rate_hz // 1000)
    * (2 * np.pi * freq / sample_rate_hz))
```

```python
sample_rate = (2*3*5*7)**2
Audio(sin_freq_ms(800, 1234), rate=sample_rate)
```

```python
L = ListeningExperiment(sin_freq_ms(400, 500), sin_freq_ms(400, 500), sample_rate)
Audio(L.place_in_sequence(200), rate=L.sample_rate)
```

```python
sound_ms = 500
offset_ms = 100

L = ListeningExperiment(sin_freq_ms(400, sound_ms),
                        np.concatenate([sin_freq_ms(0, offset_ms), sin_freq_ms(400, sound_ms - offset_ms)]),
                        sample_rate)
Audio(L.place_in_sequence(200), rate=L.sample_rate)
```
